#### lets import our libraries

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.tree import export_text
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
```

### We can specify properties and target value using sklearn

```python
# return x_y cna split our data
# as frame = show data like pandas
x , y = load_breast_cancer(return_X_y= True ,as_frame= True)
x
```

```python
y
```

### in this Dataset our target is: diagnosis
### and our featurs are: radius_mean میانگین شعاع
### texture_mean میانگین بافت
### in target column we can see M and B:
### M = malignant بد خیم
### B = benign 

## Split data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.33, shuffle= False)
x_train
```

## Modeling

### We can change the depth of data review

```python
# This reduces the confidence level:
dtc1 = DecisionTreeClassifier(max_depth= 3,) # class_weight= {0:0.30, 1:0.7}
dtc1.fit(x_train, y_train)
```

```python
plot_tree(dtc1, filled= True, fontsize= 6)
plt.plot()
```

### Tree display with text

```python
print(export_text(dtc1, feature_names= x.columns))
```

## Test the model

```python
dtc1.predict(x_test)
```

```python
dtc1.predict_proba(x_test)
```

## acuuracy 

```python
y_pre = dtc1.predict(x_test)
metrics.accuracy_score(y_test, y_pre)
```

```python
y_pre1 = dtc1.predict(x_test)
metrics.accuracy_score(y_test, y_pre)
```

### But we have an unbalanced dataset and we have to use Confusion matrix

### Precisoion

```python
metrics.precision_score(y_test, y_pre)
```

### Recall

```python
metrics.recall_score(y_test, y_pre)
```

```python
metrics.confusion_matrix(y_test, y_pre)
```
